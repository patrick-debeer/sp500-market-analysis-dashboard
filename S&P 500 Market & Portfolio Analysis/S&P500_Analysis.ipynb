```python
import yfinance as yf
import pandas as pd
import requests
from io import StringIO


url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"

headers = {
    "User-Agent": "Mozilla/5.0"
}

response = requests.get(url, headers=headers)

# Read HTML tables from the page content
table = pd.read_html(StringIO(response.text))[0]

# Convert ticker format for Yahoo Finance
tickers = table['Symbol'].str.replace('.', '-', regex=False).tolist()

print(f"Total tickers pulled: {len(tickers)}")
print(tickers[:10])
```

```python
raw_data_2023 = yf.download(tickers, start='2023-01-01', end='2023-12-31', auto_adjust=True)
print(raw_data_2023.shape)
print(raw_data_2023.head())
```

```python
raw_data_2024 = yf.download(tickers, start='2024-01-01', end='2024-12-31', auto_adjust=True)
raw_data_2025 = yf.download(tickers, start='2025-01-01', end='2025-12-31', auto_adjust=True)
print(raw_data_2025.shape)
print(raw_data_2025.head())
print(raw_data_2024.shape)
print(raw_data_2024.head())
```

```python
import datetime
today = datetime.date.today().strftime('%Y-%m-%d')
# 2026 batch - always pulls to today
raw_data_2026 = yf.download(tickers, start='2026-01-01', end=today, auto_adjust=True)
print(f"Downloaded 2026 data up to {today}")
print(raw_data_2026.shape)
print(raw_data_2026.shape)
print(raw_data_2026.head())
```

```python
raw_data_all = pd.concat([raw_data_2023, raw_data_2024, raw_data_2025, raw_data_2026])
print(f"Total rows across all years: {raw_data_all.shape}")
```

```python
close_prices = raw_data_all['Close'].copy()

df = close_prices.reset_index().melt(
    id_vars='Date',
    var_name='Ticker',
    value_name='Close_Price'
)

df = df.dropna(subset=['Close_Price'])
df['Date'] = pd.to_datetime(df['Date'])
df['Close_Price'] = df['Close_Price'].round(2)
df = df.sort_values(['Ticker', 'Date']).reset_index(drop=True)

print(df.shape)
print(df.head(10))
```

```python
sector_df = table[['Symbol', 'GICS Sector']].copy()
sector_df.columns = ['Ticker', 'Sector']
sector_df['Ticker'] = sector_df['Ticker'].str.replace('.', '-', regex=False)

df = df.merge(sector_df, on='Ticker', how='left')

print(df.head(10))
print(df['Sector'].value_counts())
```

```python
df = df.sort_values(['Ticker', 'Date'])

df['Daily_Return'] = df.groupby('Ticker')['Close_Price'].transform(
    lambda x: x.pct_change().round(4)
)

df['MA_30'] = df.groupby('Ticker')['Close_Price'].transform(
    lambda x: x.rolling(window=30).mean().round(2)
)

df['Volatility_30'] = df.groupby('Ticker')['Daily_Return'].transform(
    lambda x: x.rolling(window=30).std().round(4)
)

df['Cumulative_Return'] = df.groupby('Ticker')['Close_Price'].transform(
    lambda x: ((x / x.iloc[0]) - 1).round(4)
)

print(df.shape)
print(df.head(10))
```

```python
### Using SQl
import sqlite3
conn = sqlite3.connect('sp500_analysis.db')

df.to_sql('stock_prices', conn, if_exists='replace', index=False)

print(f"Saved successfully — {len(df)} rows")

verify = pd.read_sql("SELECT * FROM stock_prices LIMIT 5", conn)
print(verify)

conn.close()
```

```python
import datetime
today = datetime.date.today().strftime('%Y-%m-%d')

sp500_index = yf.download('^GSPC', start='2023-01-01', end=today, auto_adjust=True)
print(f"Downloaded S&P 500 data up to {today}")
print(sp500_index.shape)
print(sp500_index.head())
```

```python
# Clean the index data
sp500_clean = sp500_index['Close'].reset_index()
sp500_clean.columns = ['Date', 'SP500_Close']
sp500_clean['Date'] = pd.to_datetime(sp500_clean['Date'])
sp500_clean['SP500_Close'] = sp500_clean['SP500_Close'].round(2)

# Calculate daily return and cumulative return for the index
sp500_clean['SP500_Daily_Return'] = sp500_clean['SP500_Close'].pct_change().round(4)
sp500_clean['SP500_Cumulative_Return'] = ((sp500_clean['SP500_Close'] / sp500_clean['SP500_Close'].iloc[0]) - 1).round(4)

print(sp500_clean.shape)
print(sp500_clean.head(10))
```

```python
# Save to SQLite as its own table
conn = sqlite3.connect('sp500_analysis.db')

sp500_clean.to_sql('sp500_index', conn, if_exists='replace', index=False)

print("S&P 500 index saved successfully")

verify = pd.read_sql("SELECT * FROM sp500_index LIMIT 5", conn)
print(verify)

conn.close()
```

```python
conn = sqlite3.connect('sp500_analysis.db')

sp500_final = pd.read_sql("""
    SELECT SP500_Cumulative_Return as SP500_Final_Return
    FROM sp500_index
    WHERE Date = (SELECT MAX(Date) FROM sp500_index)
""", conn).iloc[0,0]

print(f"S&P 500 total return 2023-2026: {round(sp500_final * 100, 2)}%")
```

```python
conn = sqlite3.connect('sp500_analysis.db')

master_summary = pd.read_sql("""
    SELECT
        sp.Ticker,
        sp.Sector,
        ROUND(sp.Cumulative_Return * 100, 2) as Total_Return_Pct,
        ROUND(sp.Volatility_30, 4) as Volatility,
        ROUND(sp.MA_30, 2) as MA_30,
        CASE 
            WHEN sp.Volatility_30 < 0.01 THEN 'Low Risk'
            WHEN sp.Volatility_30 BETWEEN 0.01 AND 0.02 THEN 'Medium Risk'
            WHEN sp.Volatility_30 > 0.02 THEN 'High Risk'
            ELSE 'Unknown'
        END as Risk_Category,
        CASE
            WHEN sp.Cumulative_Return > 0.20 THEN 'Strong Performer'
            WHEN sp.Cumulative_Return BETWEEN 0 AND 0.20 THEN 'Moderate Performer'
            WHEN sp.Cumulative_Return < 0 THEN 'Underperformer'
            ELSE 'Unknown'
        END as Performance_Category,
        CASE
            WHEN sp.Cumulative_Return > si.SP500_Cumulative_Return THEN 'Beat Market'
            ELSE 'Below Market'
        END as VS_Market,
        RANK() OVER (ORDER BY sp.Cumulative_Return DESC) as Overall_Rank,
        RANK() OVER (PARTITION BY sp.Sector ORDER BY sp.Cumulative_Return DESC) as Sector_Rank
    FROM stock_prices sp
    JOIN sp500_index si ON sp.Date = si.Date
    WHERE sp.Date = (SELECT MAX(Date) FROM stock_prices)
""", conn)

print(master_summary.shape)
print(master_summary.head(10))

conn.close()
```

```python
conn = sqlite3.connect('sp500_analysis.db')

master_summary.to_sql('master_summary', conn, if_exists='replace', index=False)

total_stocks = len(master_summary)
beat_market = len(master_summary[master_summary['VS_Market'] == 'Beat Market'])
print(f"Total stocks: {total_stocks}")
print(f"Stocks that beat the S&P 500: {beat_market} ({round(beat_market/total_stocks*100, 1)}%)")
print(f"\nPerformance breakdown:")
print(master_summary['Performance_Category'].value_counts())
print(f"\nRisk breakdown:")
print(master_summary['Risk_Category'].value_counts())
print(f"\nSector breakdown:")
print(master_summary.groupby('Sector')['Total_Return_Pct'].mean().round(2).sort_values(ascending=False))

conn.close()
```

```python
# ============================================
# PORTFOLIO ANALYZER
# ============================================
import datetime

# Set up sp variable for portfolio calculations
sp = sp500_clean.copy()
sp['Date'] = pd.to_datetime(sp['Date'])
print("S&P 500 data ready for portfolio calculations")
print(f"Date range: {sp['Date'].min().date()} to {sp['Date'].max().date()}")
print(sp.columns.tolist())
```

```python
portfolio = [
    {"ticker": "AMZN", "shares": 25, "buy_date": "2023-02-23"},
    {"ticker": "EME",  "shares": 15, "buy_date": "2024-02-22"},
    {"ticker": "MCK",  "shares": 8,  "buy_date": "2025-12-30"},
    {"ticker": "MU",   "shares": 17, "buy_date": "2026-03-10"},
    {"ticker": "NVDA", "shares": 9,  "buy_date": "2026-03-23"},
    {"ticker": "NVDA", "shares": 21, "buy_date": "2024-08-06"},
]
```

```python
results = []

for holding in portfolio:
    ticker = holding['ticker']
    shares = holding['shares']
    buy_date = pd.to_datetime(holding['buy_date'])
    
    stock_df = df[df['Ticker'] == ticker].sort_values('Date')
    
    if len(stock_df) == 0:
        print(f"WARNING: {ticker} not found in dataset. Skipping.")
        continue
    
    min_date = stock_df['Date'].min()
    max_date = stock_df['Date'].max()
    
    if buy_date < min_date:
        print(f"WARNING: {ticker} buy date {buy_date.date()} before data start. Using {min_date.date()}.")
        buy_date = min_date
    elif buy_date > max_date:
        print(f"WARNING: {ticker} buy date {buy_date.date()} after data end. Skipping.")
        continue
    
    available = stock_df[stock_df['Date'] >= buy_date]
    if len(available) == 0:
        print(f"WARNING: No data for {ticker} after {buy_date.date()}. Skipping.")
        continue
    actual_buy_date = available['Date'].iloc[0]
    
    buy_price = stock_df[stock_df['Date'] == actual_buy_date]['Close_Price'].values[0]
    current_price = stock_df['Close_Price'].iloc[-1]
    
    amount_invested = shares * buy_price
    current_value = shares * current_price
    total_gain = current_value - amount_invested
    total_return_pct = ((current_price - buy_price) / buy_price) * 100
    
    sp_on_buy = sp[sp['Date'] >= actual_buy_date]['SP500_Close'].iloc[0]
    sp_current = sp['SP500_Close'].iloc[-1]
    sp_shares_equiv = amount_invested / sp_on_buy
    sp_current_value = sp_shares_equiv * sp_current
    sp_return_pct = ((sp_current - sp_on_buy) / sp_on_buy) * 100
    
    vs_market_pct = total_return_pct - sp_return_pct
    vs_market_dollars = current_value - sp_current_value
    
    results.append({
        'Ticker': ticker,
        'Shares': shares,
        'Buy_Date': actual_buy_date.date(),
        'Buy_Price': round(buy_price, 2),
        'Amount_Invested': round(amount_invested, 2),
        'Current_Price': round(current_price, 2),
        'Current_Value': round(current_value, 2),
        'Total_Gain': round(total_gain, 2),
        'Total_Return_Pct': round(total_return_pct, 2),
        'SP500_Return_Pct': round(sp_return_pct, 2),
        'VS_Market_Pct': round(vs_market_pct, 2),
        'VS_Market_Dollars': round(vs_market_dollars, 2),
        'Beat_Market': 'Yes' if vs_market_pct > 0 else 'No',
        'Purchase_ID': f"{ticker}_{actual_buy_date.date()}"
    })

portfolio_df = pd.DataFrame(results)
print(portfolio_df.to_string())
```

```python
total_invested = portfolio_df['Amount_Invested'].sum()
total_value = portfolio_df['Current_Value'].sum()
total_gain = portfolio_df['Total_Gain'].sum()
total_return = ((total_value - total_invested) / total_invested) * 100

sp_equiv_value = portfolio_df.apply(
    lambda r: (r['Amount_Invested'] / sp[sp['Date'] >= pd.to_datetime(r['Buy_Date'])]['SP500_Close'].iloc[0]) * sp['SP500_Close'].iloc[-1],
    axis=1
).sum()
sp_equiv_return = ((sp_equiv_value - total_invested) / total_invested) * 100
vs_market_total = total_return - sp_equiv_return

print("=" * 40)
print("PORTFOLIO SUMMARY")
print("=" * 40)
print(f"Total invested:      ${total_invested:,.2f}")
print(f"Current value:       ${total_value:,.2f}")
print(f"Total gain:          ${total_gain:,.2f}")
print(f"Portfolio return:    {total_return:.2f}%")
print(f"S&P 500 return:      {sp_equiv_return:.2f}%")
print(f"vs Market:           {vs_market_total:.2f}%")
print("=" * 40)

portfolio_df.to_csv(r'C:\Users\patri\portfolio_analysis.csv', index=False)

summary_df = pd.DataFrame([{
    'Total_Invested': round(total_invested, 2),
    'Current_Value': round(total_value, 2),
    'Total_Gain': round(total_gain, 2),
    'Total_Return_Pct': round(total_return, 2),
    'SP500_Return_Pct': round(sp_equiv_return, 2),
    'VS_Market_Pct': round(vs_market_total, 2),
    'VS_Market_Dollars': round(total_value - sp_equiv_value, 2)
}])
summary_df.to_csv(r'C:\Users\patri\portfolio_summary.csv', index=False)

print("\nSaved portfolio_analysis.csv and portfolio_summary.csv to C:\\Users\\patri")
```

```python
# Rebuild stock_prices with returns calculated from each stock's first appearance
# but for portfolio stocks, reset from their purchase date

portfolio_dates = {
    'AMZN': '2023-02-23',
    'EME': '2024-02-22',
    'MCK': '2025-12-30',
    'MU': '2026-03-10',
    'NVDA': '2024-08-06',
    'S&P 500': '2023-02-23'
}

portfolio_returns = []

for ticker, start_date in portfolio_dates.items():
    start_date = pd.to_datetime(start_date)
    ticker_df = df_with_sp500[df_with_sp500['Ticker'] == ticker].copy()
    ticker_df = ticker_df[ticker_df['Date'] >= start_date].sort_values('Date')
    
    if len(ticker_df) == 0:
        continue
        
    base_price = ticker_df['Close_Price'].iloc[0]
    ticker_df['Portfolio_Return'] = ((ticker_df['Close_Price'] / base_price) - 1) * 100
    portfolio_returns.append(ticker_df[['Date', 'Ticker', 'Close_Price', 'Portfolio_Return', 'Sector']])

portfolio_prices = pd.concat(portfolio_returns, ignore_index=True)
portfolio_prices.to_csv(r'C:\Users\patri\portfolio_prices.csv', index=False)
print(f"Exported {len(portfolio_prices)} rows")
print(portfolio_prices.head())
```

```python
# Fast vectorized version
portfolio_holdings = [
    {"ticker": "AMZN", "shares": 25, "buy_date": "2023-02-23"},
    {"ticker": "EME",  "shares": 15, "buy_date": "2024-02-22"},
    {"ticker": "MCK",  "shares": 8,  "buy_date": "2025-12-30"},
    {"ticker": "MU",   "shares": 17, "buy_date": "2026-03-10"},
    {"ticker": "NVDA", "shares": 9,  "buy_date": "2026-03-23"},
    {"ticker": "NVDA", "shares": 21, "buy_date": "2024-08-06"},
]

# Build one dataframe per holding with daily value
holding_dfs = []

for holding in portfolio_holdings:
    ticker = holding['ticker']
    shares = holding['shares']
    buy_date = pd.to_datetime(holding['buy_date'])
    
    # Get this stock's prices from buy date onward
    stock = df[(df['Ticker'] == ticker) & (df['Date'] >= buy_date)].copy()
    
    if len(stock) == 0:
        continue
    
    # Buy price is first available price
    buy_price = stock['Close_Price'].iloc[0]
    
    stock['Holding_Value'] = shares * stock['Close_Price']
    stock['Amount_Invested'] = shares * buy_price
    stock['Purchase_ID'] = f"{ticker}_{buy_date.date()}"
    
    holding_dfs.append(stock[['Date', 'Holding_Value', 'Amount_Invested']])

# Combine all holdings — for each date sum all active holdings
combined = pd.concat(holding_dfs)
portfolio_daily = combined.groupby('Date').agg(
    Portfolio_Value=('Holding_Value', 'sum'),
    Total_Invested=('Amount_Invested', 'sum')
).reset_index()

portfolio_daily['Portfolio_Return_Pct'] = (
    (portfolio_daily['Portfolio_Value'] - portfolio_daily['Total_Invested']) 
    / portfolio_daily['Total_Invested'] * 100
).round(4)

# Add S&P 500 benchmark from earliest buy date
sp_base_date = pd.to_datetime('2023-02-23')
sp_filtered = sp500_clean[sp500_clean['Date'] >= sp_base_date].copy()
sp_base_price = sp_filtered['SP500_Close'].iloc[0]
sp_filtered['SP500_Return_Pct'] = ((sp_filtered['SP500_Close'] / sp_base_price) - 1) * 100

portfolio_daily = portfolio_daily.merge(
    sp_filtered[['Date', 'SP500_Return_Pct']],
    on='Date',
    how='left'
)

portfolio_daily.to_csv(r'C:\Users\patri\portfolio_daily.csv', index=False)
print(f"Exported {len(portfolio_daily)} rows")
print(portfolio_daily.tail(5))
```
